```python
import matplotlib.pyplot as plt # this helps with writing cleaner code
import matplotlib as mpl
import random # used to randomly generate values

import numpy as np # useful for holding data for plotting

np.random.seed(42) # set the seed for random number generation to get the same values each time
groups = ["a", "b", "c"]
group_data =[random.choice(groups) for x in range(50)]
var_data = np.random.randn(50) # generate 50 random floats
int_data = np.random.randint(0, 50, 50) # generate 50 random integers between 0 and 50
plot_data = {"groups": group_data, "vars": var_data, "counts": int_data}
```

```python
import seaborn as sbn
import pandas as pd
```

```python
plot_df = pd.DataFrame(plot_data) # seaborn uses a dataframe as a common input data format
plot_df.head()
```

```python
sbn.boxplot(plot_df, x="groups", y="vars", hue="groups")
plt.show() #make it a habit of including plt.sh
```

```python
facet = sbn.FacetGrid(plot_df, col="groups", hue="groups")
facet.map(sbn.scatterplot, "vars", "counts")
plt.show()
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(6, 3), layout="constrained")
sbn.boxplot(plot_df, x="groups", y="vars", hue="groups", ax=ax[1])
sbn.kdeplot(plot_df, x="counts", hue="groups", ax=ax[0]) #kdeplot= kernel distribution plot
plt.show()
```

```python
fig, ax = plt.subplot_mosaic([["box", "scatter"],["hist", "hist"]], #box=top left, scatter=right and histograms at the buttom
                             figsize=(7, 7), layout="constrained")
sbn.boxplot(plot_df, x="groups", y="vars", hue="groups", ax=ax["box"]) #only in subplot mosaic can you use strings for the axes. box=axes 1
sbn.scatterplot(plot_df, x="counts", y="vars", hue="groups", ax=ax["scatter"]) #scatter= graph 2
sbn.kdeplot(plot_df, x="counts", hue="groups", ax=ax["hist"])#hist=graph 3
ax["box"].set_yscale("log")
plt.show()
```
